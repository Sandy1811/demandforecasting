```python
# STEP #3: PREPARE THE DATA FOR TRAINING/ DATA CLEANING
X = data.drop(['Names', 'emails', 'Country','Clicked'], axis=1).values

X
```

```python
y = data["Clicked"].values

y
```

```python
# Features Scaling

from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X1 = sc.fit_transform(X)
```

```python
X1
```

```python
# Dividing the dataset into Training and testing set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X1, y , test_size = 0.25)
```

```python
X_train
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
# Model Training
# Fitting Logistic Regression to the Training set
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(random_state = 0)
classifier.fit(X_train, y_train)
```

```python
# STEP#5: MODEL TESTING
y_pred = classifier.predict_proba(X_train)
y_pred
```

```python
y_pred_train = classifier.predict(X_train)
y_pred_train
```

```python
# Classification metrics to measure the performance of the model

from sklearn.metrics import classification_report, confusion_matrix
cm = confusion_matrix(y_train, y_pred_train)
sns.heatmap(cm, annot= True, fmt="d")
```

```python
y_predict_test = classifier.predict(X_test)
y_predict_test
```

```python
cm = confusion_matrix(y_test, y_predict_test)
sns.heatmap(cm, annot=True, fmt="d")
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_predict_test))
```
